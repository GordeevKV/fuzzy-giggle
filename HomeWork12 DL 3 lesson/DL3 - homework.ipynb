```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Flatten, Input

# Загрузка данных
train_data = pd.read_csv('C:/Users/285/Desktop/data_science/HomeWork12 DL 3 lesson/train.csv')
test_data = pd.read_csv('C:/Users/285/Desktop/data_science/HomeWork12 DL 3 lesson/test.csv')

# Предобработка данных
X_train = train_data.drop('label', axis=1).values
y_train = train_data['label'].values

# Измените форму X_train на (количество образцов, 28, 28)
X_train = X_train.reshape(-1, 28, 28)

# Создание модели
model = Sequential()
model.add(Input(shape=(28, 28)))  # Используйте Input для указания формы
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))

# Компиляция модели
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Обучение модели
model.fit(X_train, y_train, epochs=10, batch_size=32)

# Предсказание на тестовых данных
X_test = test_data.values.reshape(-1, 28, 28)  # Измените форму тестовых данных
predictions = model.predict(X_test)
predicted_labels = predictions.argmax(axis=1)

# Создание файла для отправки
submission = pd.DataFrame({'ImageId': range(1, len(predicted_labels) + 1), 'Label': predicted_labels})
submission.to_csv('submission.csv', index=False)

print("Файл 'submission.csv' создан успешно!")
```
