# Задание 1: Химический анализ родника

```python
import pandas as pd
import re
from IPython.display import display

# Загрузка данных из файла
file_path = 'C:/Users/285/Desktop/data_science/HomeWork4/HIM.csv'
df = pd.read_csv(file_path, sep=';')

# Проверка соответствия нормативу
def norm_check(value, norm):
    limits = re.findall(r'[\d\.]+', norm)
    if 'не более' in norm:
        return value <= float(limits[0]), f'не более {limits[0]}'
    elif 'в пределах' in norm:
        return float(limits[0]) <= value <= float(limits[1]), f'в пределах {limits[0]}-{limits[1]}'
    return False, 'неизвестный норматив'

# Обработка данных
df['Результат анализа'] = pd.to_numeric(df['Результат анализа'], errors='coerce')
df['Вывод'] = df.apply(lambda x: f"Результат анализа {x['Результат анализа']} {x['Единица измерений']} при нормативе {x['Норматив']}. " +
                           ('Результат в норме.' if norm_check(x['Результат анализа'], x['Норматив'])[0] else 'Результат не в норме.'),
                       axis=1)

# Отображение результата
df.set_index('Показатель', inplace=True)
display(df)

# Сохранение результата
output_path = 'C:/Users/285/Desktop/data_science/HomeWork4/ВЫВОДЫ Химический анализ родника в Нескучном саду.xlsx'
df.to_excel(output_path)
```
