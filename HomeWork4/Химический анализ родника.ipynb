# Задание 1: Химический анализ родника

```python
import pandas as pd
import re

# Загрузка данных из файла
file_path = 'C:/Users/285/Desktop/data_science/HomeWork4/HIM.csv'
df = pd.read_csv(file_path, sep=';')

# Функция для обработки норматива
def check_normative(row):
    result = row['Результат анализа']
    normative = row['Норматив']
    
    if isinstance(result, str) or isinstance(normative, str):
        # Если результат или норматив не числовые (например, "б/цвета"), вернуть строку с описанием
        return f'{row["Показатель"]}: "{result}"'

    if 'в пределах' in normative:
        # Пример: в пределах 6-9
        match = re.search(r'в пределах (\d+)-(\d+)', normative)
        if match:
            lower, upper = map(float, match.groups())
            if lower <= result <= upper:
                return f'результат анализа в норме'
            else:
                return f'результат анализа не в норме'
    
    elif 'не более' in normative:
        # Пример: не более 2-3
        match = re.search(r'не более (\d+)', normative)
        if match:
            upper = float(match.group(1))
            if result <= upper:
                return f'результат анализа в норме'
            else:
                return f'результат анализа не в норме'
    
    return 'результат анализа неизвестен'

# Применение функции к DataFrame
df['Вывод'] = df.apply(check_normative, axis=1)
df.set_index('Показатель', inplace=True)

# Вывод итогового DataFrame
print(df)
```
